```python
import pandas as pd
```

```python
gptData = pd.read_csv('/content/drive/MyDrive/datasets/ChatGPT.csv')
```

```python
gptData2 = pd.read_csv('/content/drive/MyDrive/datasets/cleaned_dataset.csv')
```

```python
# renaming the column labels.
gptData2.rename(columns={'tweet_created': 'Date'}, inplace=True)
gptData2.rename(columns={'text': 'Tweet'}, inplace=True)
gptData2.rename(columns={'user_username': 'User'}, inplace=True)
gptData2.rename(columns={'user_location': 'Location'}, inplace=True)
gptData2.rename(columns={'user_description': 'Description'}, inplace=True)
gptData2.rename(columns={'user_created': 'UserCreated'}, inplace=True)
gptData2.rename(columns={'user_followers_count': 'UserFollowers'}, inplace=True)
gptData2.rename(columns={'user_verified': 'UserVerified'}, inplace=True)
gptData2.rename(columns={'retweet_count': 'Retweets'}, inplace=True)
gptData2.rename(columns={'like_count': 'Likes'}, inplace=True)
gptData2.rename(columns={'tweet_id': 'Unnamed: 0'}, inplace=True)
```

```python
# drop - user_id, user_name, user_tweet_count, user_following_count, source, reply_count, impression_count
gptData2.drop(columns=['tweet_extracted'], inplace=True)
gptData2.drop(columns=['user_id'], inplace=True)
gptData2.drop(columns=['user_name'], inplace=True)
gptData2.drop(columns=['user_tweet_count'], inplace=True)
gptData2.drop(columns=['user_following_count'], inplace=True)
gptData2.drop(columns=['source'], inplace=True)
gptData2.drop(columns=['impression_count'], inplace=True)
gptData2.drop(columns=['reply_count'], inplace=True)
gptData2.drop(columns=['lang'], inplace=True)
```

```python
merged_dataset = pd.concat([gptData, gptData2], axis=0, ignore_index=True) # dataset merged
```

```python
# checking if data is merged properly.
# Printing the first few rows of dataset1's 'Unnamed: 0' column
print("Dataset 1 'Unnamed: 0' column:")
print(gptData['Unnamed: 0'].head())

# Printing the last few rows of dataset1's 'Unnamed: 0' column
print("\nDataset 1 'Unnamed: 0' column:")
print(gptData['Unnamed: 0'].tail())

# Printing the first few rows of dataset2's 'Unnamed: 0' column
print("\nDataset 2 'Unnamed: 0' column:")
print(gptData2['Unnamed: 0'].head())

# Printing the first few rows of dataset2's 'Unnamed: 0' column
print("\nDataset 2 'Unnamed: 0' column:")
print(gptData2['Unnamed: 0'].tail())

# Printing the first few rows of the merged 'Unnamed: 0' column
print("\nMerged 'Unnamed: 0' column:")
print(merged_dataset['Unnamed: 0'].head())

# Printing the last few rows of the merged 'Unnamed: 0' column
print("\nMerged 'Unnamed: 0' column:")
print(merged_dataset['Unnamed: 0'].tail())
```
